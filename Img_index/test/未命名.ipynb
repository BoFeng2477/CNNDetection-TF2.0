```python
import pandas as pd
```

```python
filename = 'whichfaceisreal_test.csv'
df = pd.read_csv(filename,usecols=['file','label'])
```

```python
df
```

```python
droplist = []
for i,row in df.iterrows():
    if row['file'].split('/')[1][0]=='.':
        droplist.append(i)
df = df.drop(droplist).reset_index(drop=True)
```

```python
df
```

```python
df.to_csv(filename,index=0)
```
